# Задание 0

Создайте функцию по нахождению уникальных элементов из двух списков



```
a = [1, 2, 3, 4]
b = [3, 4, 5, 6]
```



````python
def unique_el(a,b):
  uni = []
  for i in a:
    if i not in b:
      uni.append(i)
  for u in b:
    if u not in a:
      uni.append(u)
  print(sorted(uni))


def main():
  a = [1, 2, 3, 4]
  b = [3, 4, 5, 6]
  unique_el(a,b)


if __name__ == '__main__':
  main()
````

# Задание 2

Рыцарь и дракон

Цель: создать небольшую игру, в которой вам необходимо играть за рыцаря и сразиться с драконом

Требования:

- Создание персонажа (имя, информация о доспехах, оружии, урон, здоровье)
- Управление персонажем и мини сюжет
- Создание дракона (Имя, информация о здоровье и уроне)

- Реализовать бой между драконом и рыцарем

````python
from random import randint, choice

def knight_creating(knight):
  qu1 = input('Итак, жил да был рыцарь в далёком королевстве. И звали его... ')
  knight['name']=qu1

  print(f'\nТочно! Звали его {qu1}.\nРешил как-то раз {qu1} подвиг совершить, да сразить дракона, докучавшего всему королевству.')
  print('Начал собираться в путь-дорогу, но ему нужно было сначла приодеться.\n')

  qu1 = int(input('На голову решил надеть шлем с защитой ... '))
  knight['armour']['hat']=qu1

  qu1 = int(input('На тулово напялил нагрудник с защитой ... '))
  knight['armour']['chest']=qu1

  qu1 = int(input('На ноги натянул хорошие штанцы с защитой ... '))
  knight['armour']['legs']=qu1

  qu1 = int(input('На ступни в спешке надел ботинки с защитой ... '))
  knight['armour']['feet']=qu1

  knight['health']+=knight['armour']['hat']+knight['armour']['chest']+knight['armour']['legs']+knight['armour']['feet']

  qu1 = int(input('Чуть не забыли про меч! А взял он неплохой палаш с максимальным уроном ... '))
  knight['damage']=qu1
  print(f'\nВот и собрался рыцарь-{knight["name"]} выдвигаться, совершать задуманное.')


def challanges(knight):
  dangers = ['Очень глубокая яма с шипами внизу','Автоматический арбалет с ядовитой стрелой','Комната, охваченная огнём']
  tridveri = ' /\   /\   /\ \n|. | |. | |. |\n|__| |__| |__|'
  answers = ''
  knightIsDead = False
  print('Подходит наш рыцарь к башне, где на самой верхушке сидит дракон.')
  height = int(input('И, собсна, видит там не много не мало - ... этажей!\n'))
  for i in range(1,height):
    answ = randint(1,3)
    answers+=str(answ)
  print(f'Сгенерирована комбинация {answers}\n')

  print('Заходит внутрь через главные врата и наблюдает перед собой 3 двери.')
  print(tridveri)

  qu2 = int(input('Какую же дверь ему выбрать? (1 или 2 или 3)'))
  if (qu2 == int(answers[0])) and (height>2) and (knightIsDead==False):
    print(f'Рыцарь зашёл в {qu2} дверь и увидел лестницу на след. этаж.')
    print('Поднявшись, он снова увидел 3 двери.')
    print('Сразу стало понятно, что это своеобразный лабиринт с лавушками.\n')
    for i in range(1,height-1):
      if (knightIsDead==False):
        qu2 = int(input('Какую же дверь ему выбрать? (1 или 2 или 3)'))
        if qu2 == int(answers[i]) and (height>2):
          print(f'Рыцарь зашёл в {qu2} дверь и увидел лестницу на след. этаж.')
          print('Поднявшись, он снова увидел 3 двери.\n')
        else:
          print(f'Рыцарь зашёл в {qu2} дверь.')
          print('Дверь быстро закрылась и затолкнула героя внутрь.')
          print(f'К сожалению его ждал(а) {choice(dangers)}. Он погиб.\n')
          knightIsDead = True
          break
    print('Наконец наш путник поднялся к дракону.')
  else:
    print(f'Рыцарь зашёл в {qu2} дверь.')
    print('Дверь быстро закрылась и затолкнула героя внутрь.')
    print(f'К сожалению его ждал(а) {choice(dangers)}. Он погиб.\n')
    knightIsDead = True

  knight['isdead']=knightIsDead
  return knight


def dragon_creating(dragon):
  qu1 = input('А предстал перед нашим рыцарем величайший дракон по прозвищу ... ')
  dragon['name']=qu1

  qu1 = int(input(f'{dragon["name"]} был довольно таки сильным существом и максимум наносил ... урона. \n'))
  dragon['damage'] = qu1


def fighting(knight,dragon):
  print('И вот наконец-то рыцарь стоит перед драконом. Но страх не охватывает ни нашего героя, ни его противника. Бой начинается!\n')
  while knight['health']>0 and dragon['health']>0:
    qu1 = int(input('Ход рыцаря. Что выберешь?\n 1. Прямой удар.\n 2. Удар исподтишка (урон х2, но нужно хорошее везение).\n'))
    if qu1 == 1:
      dmg = randint(1,knight['damage'])
      dragon['health']-=dmg
      print(f'Рыцарь наносит прямой удар - {dmg} урона. У дракона осталось {dragon["health"]}\n')
    elif qu1 == 2:
      chc = choice([True,False])
      if chc == True:
        dmg = 2*randint(1,knight['damage'])
        dragon['health']-=dmg
        print(f'Вы попытались ударить исподтишка и .... Вам повезло! Рыцарь наносит двойной урон - {dmg}. У дракона осталось {dragon["health"]} ХП\n')
      else:
        dmg = 1
        dragon['health']-=dmg
        print(f'Вы попытались ударить исподтишка и .... Вам не повезло)))) Рыцарь наносит {dmg} урон. У дракона осталось {dragon["health"]} ХП\n')

    print('Ход дракона')
    dragonchoice = choice([1,2])
    if dragonchoice==1:
      dmg = randint(1,dragon['damage'])
      knight['health']-=dmg
      print(f'Дракон выбрал прямой удар. Он наносит рыцарю {dmg}. У вас осталось {knight["health"]} ХП\n')
    elif dragonchoice == 2:
      chc = choice([True,False])
      if chc == True:
        dmg = 2*randint(1,dragon['damage'])
        dragon['health']-=dmg
        print(f'Дракон попытался подловить вас и ....  Ему повезло! Дракон наносит двойной урон - {dmg}. У рыцаря осталось {knight["health"]} ХП\n')
      else:
        dmg = 1
        dragon['health']-=dmg
        print(f'Дракон попытался подловить вас и ....  Ему не повезло! Дракон не наносит вам урона. У рыцаря всё ещё {knight["health"]} ХП\n')



def main():
  knight = {'name':'',
            'armour':{'hat':'','chest':'','legs':'','feet':''},
            'damage':'',
            'health':10,
            'isdead':False}
  dragon = {'name':'','health':100,'damage':'','isdead':False}

  qu = input('Ну что, начнём нашу историю? (Да или Нет) - ')
  if qu == 'Да':
    knight_creating(knight)
    if challanges(knight)['isdead']==False:
      dragon_creating(dragon)
    else:
      print('Рыцарь не дошел до дракона. Он умер, выбрав не ту дверь...')

    if knight['isdead']==False:
      fighting(knight,dragon)

    if knight['health']>0 and knight['isdead']==False:
      print(f'{dragon["name"]} пал! Победа за рыцарем! КОНЕЦ!')
    else:
      print(f'{knight["name"]} погиб... Победа за Драконом. КОНЕЦ!')

  else:
    print('Ну чтож, тогда в следующий раз!')

if __name__ == '__main__':
  main()
````

# Задание 3


Цель - создать менеджера команды Pokémon, который позволит пользователям:

- Добавлять покемонов в свою команду. (если такого покемона еще нет в команде)
- Удалять покемонов из их команды.
- Просматривать подробную информацию обо всех покемонах в команде.
- Находить покемона по имени.
- Устраивать тренировочный бой между двумя покемонами

Для данной задачи используйте: https://pokeapi.co/


````python
import requests


def get_pokemon_data(name):
    url = f"https://pokeapi.co/api/v2/pokemon/{name.lower()}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Покемон '{name}' не найден в PokeAPI.")
        return None


def pokemon_info(pokemon_data):
    if pokemon_data:
        types = ", ".join([t['type']['name'] for t in pokemon_data['types']])
        abilities = ", ".join([a['ability']['name'] for a in pokemon_data['abilities']])
        print(f"Имя: {pokemon_data['name'].title()}\nТип: {types}\nСпособности: {abilities}")
    else:
        print("Покемон не найден.")


def add_pokemon(team, name):
    if name.title() not in [p['name'].title() for p in team]:
        pokemon_data = get_pokemon_data(name)
        if pokemon_data:
            team.append(pokemon_data)
            print(f"{name.title()} добавлен в команду!")
        else:
            print(f"Не удалось добавить {name.title()} в команду.")
    else:
        print(f"{name.title()} уже в команде.")


def remove_pokemon(team, name):
    for pokemon in team:
        if pokemon['name'].title() == name.title():
            team.remove(pokemon)
            print(f"{name.title()} удален из команды.")
            return
    print(f"{name.title()} не найден в команде.")


def view_team(team):
    if team:
        print("Ваша команда:")
        for pokemon in team:
            print(pokemon_info(pokemon))
            print("-" * 20)
    else:
        print("Ваша команда пуста.")

def find_pokemon(team, name):
    for pokemon in team:
        if pokemon['name'].title() == name.title():
            print(pokemon_info(pokemon))
            return
    print(f"{name.title()} не найден в команде.")


def training_battle(team, pokemon1_name, pokemon2_name):
    pokemon1 = None
    pokemon2 = None

    for pokemon in team:
        if pokemon['name'].title() == pokemon1_name.title():
            pokemon1 = pokemon
        if pokemon['name'].title() == pokemon2_name.title():
            pokemon2 = pokemon

    if not pokemon1 or not pokemon2:
        print("Один или оба покемона не найдены в команде.")
        return

    print(f"--- Тренировочный бой: {pokemon1['name'].title()} против {pokemon2['name'].title()} ---")
    pokemon1_stats = sum(s['base_stat'] for s in pokemon1['stats'])
    pokemon2_stats = sum(s['base_stat'] for s in pokemon2['stats'])

    if pokemon1_stats > pokemon2_stats:
        print(f"{pokemon1['name'].title()} побеждает!")
    elif pokemon2_stats > pokemon1_stats:
        print(f"{pokemon2['name'].title()} побеждает!")
    else:
        print("Ничья!")


def main():
  team = []
  while True:
    qu1 = input('Что делаем?\n 1. Добавить покемона в команду.\n 2. Удалить покемона из команды.\n 3. Просмотреть команду.\n 4. Найти покемона по имени.\n 5. Тренировочный бой.\n 6. Ничего.\n') # Get input here

    if qu1 == '6':
      break

    if qu1=='1':
      name = input('Введите имя покемона: ')
      add_pokemon(team,name)
    elif qu1=='2':
      name = input('Введите имя покемона: ')
      remove_pokemon(team,name)
    elif qu1=='3':
      view_team(team)
    elif qu1=='4':
      name = input('Введите имя покемона: ')
      find_pokemon(team,name)
    elif qu1=='5':
      pokemon1_name = input('Введите имя первого покемона: ')
      pokemon2_name = input('Введите имя второго покемона: ')
      training_battle(team, pokemon1_name, pokemon2_name)


if __name__ == '__main__':
  main()
````
